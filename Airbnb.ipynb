```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for better aesthetics
sns.set_style("whitegrid")

# Create a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='price', data=df)
plt.title('Price Distribution by Room Type', fontsize=14)
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.ylim(0, df['price'].quantile(0.95))  # Limit y-axis to 95th percentile to handle outliers
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'rating' to numeric, handling non-numeric values
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Calculate mean rating per neighbourhood group
avg_rating = df.groupby('neighbourhood_group')['rating'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='neighbourhood_group', y='rating', data=avg_rating, palette='pastel')
plt.title('Average Rating by Neighbourhood Group', fontsize=14)
plt.xlabel('Neighbourhood Group', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.ylim(0, 5)  # Assuming ratings are on a 0-5 scale
plt.xticks(rotation=45)
plt.show()
```

```python
# Create a violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='neighbourhood_group', y='availability_365', data=df, palette='muted')
plt.title('Availability Distribution by Neighbourhood Group', fontsize=14)
plt.xlabel('Neighbourhood Group', fontsize=12)
plt.ylabel('Availability (Days per Year)', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
